```python
import json

from korean_romanizer.romanizer import Romanizer
```

# **Step 1: Data Preparation**

```python

# Korean text
korean_text = " 저는 David Gallo입니다"

# Romanize the text
romanized_text = Romanizer(korean_text)

print(romanized_text.romanize())
```

```python
# Path to the JSON file
json_file_path = "data/korean_data/train.json"

# Lists to store Korean and English sentences
korean_list = []
english_list = []

# Read data from the JSON file
with open(json_file_path, 'r', encoding='utf-8') as file:
    data = file.readlines()
    for line in data:
        # Parse JSON data
        json_data = json.loads(line)

        # Append Korean and English sentences to their respective lists
        korean_list.append(json_data['korean'])
        english_list.append(json_data['english'])

print(korean_list[:5])
print(english_list[:5])
print(len(korean_list))
```

```python
# Romanization
romanized_list = []
i = 0
for sentence in korean_list:
  i += 1
  romanized_text = Romanizer(sentence)
  text = romanized_text.romanize()
  romanized_list.append(text)
  # print(i)

print(romanized_list[:5])
```
