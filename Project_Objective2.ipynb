```python
import json
import csv
from korean_romanizer.romanizer import Romanizer
```

# **Step 1: Data Preparation**

```python

# Korean text
korean_text = " 저는 David Gallo입니다"

# Romanize the text
romanized_text = Romanizer(korean_text)

print(romanized_text.romanize())
```

```python
# Path to the JSON file
json_file_path = "data/korean_data/train.json"

# Lists to store Korean and English sentences
korean_list = []
english_list = []

# Read data from the JSON file
with open(json_file_path, 'r', encoding='utf-8') as file:
    data = file.readlines()
    for line in data:
        # Parse JSON data
        json_data = json.loads(line)

        # Append Korean and English sentences to their respective lists
        korean_list.append(json_data['korean'])
        english_list.append(json_data['english'])

print(korean_list[:5])
print(english_list[:5])
print(len(korean_list))
```

```python
# Romanization
romanized_list = []
i = 0
for sentence in korean_list:
  i += 1
  try:
    romanized_text = Romanizer(sentence)
    text = romanized_text.romanize()
  except:
    continue
  romanized_list.append(text)
  # print(i)

print(romanized_list[:5])
```

```python
print(len(romanized_list))
```

```python
# New DataSet
romanization_file = "data/korean_data/new_romanized_korean_english_new.csv"

with open(romanization_file, 'w', newline='', encoding = 'utf-8') as file:
  writer = csv.writer(file)
  i = 0
  for sentence in romanized_list:
    writer.writerow([korean_list[i], romanized_list[i], english_list[i]])
    i+=1
    if(i%50 == 0):
        print(f"Doing sentence {i}")
```

```python
# Path to the JSON file
json_file_path = "data/korean_data/Sentiment_Korean.tsv"

# Lists to store Korean and English sentences
korean_list = []
sentiment_list = []

# Read data from the JSON file
with open(json_file_path, 'r', encoding='utf-8', newline="" ) as file:
    data = csv.reader(file, delimiter="\t")
    for line in data:
        # Append Korean and English sentences to their respective lists
        korean_list.append(line[1])
        sentiment_list.append(line[0])

print(korean_list[:5])
print(sentiment_list[:5])
print(len(korean_list))
```

```python
# Romanization
romanized_list_sentiment = []
i = 0
for sentence in korean_list:
  i += 1
  try:
    romanized_text = Romanizer(sentence)
    text = romanized_text.romanize()
  except:
    continue
  romanized_list_sentiment.append(text)
  # print(i)

print(romanized_list_sentiment[:5])
print(len(romanized_list_sentiment))
```

```python
# New DataSet
romanization_file = "data/korean_data/new_romanized_korean_sentiment_new.csv"

with open(romanization_file, 'w', newline='', encoding = 'utf-8') as file:
  writer = csv.writer(file)
  i = 0
  for sentence in romanized_list_sentiment:
    writer.writerow([korean_list[i], romanized_list_sentiment[i], sentiment_list[i]])
    i+=1
    if(i%50 == 0):
        print(f"Doing sentence {i}")
```
